```python
import pandas as pd
import numpy as np
import matplotlib
matplotlib.use("nbagg")
import matplotlib.pyplot as plt
```

```python
mod1 = pd.DataFrame.from_csv("/media/DataShare/Taylor/SRS/modis_srs_red.csv")
mod2 = pd.DataFrame.from_csv("/media/DataShare/Taylor/SRS/modis_srs_nir.csv")
s = pd.DataFrame.from_csv("/media/DataShare/Taylor/SRS/sentinel_srs_red_nir.csv")

sen = pd.DataFrame(index=s.index)
sen['red'] = s.S2A_SR_AV_B4
sen['nir']= s.S2A_SR_AV_B8
s = None

pla = pd.DataFrame.from_csv("/media/DataShare/Taylor/SRS/planet_srs_red_nir.csv")
```

```python
fig, ax = plt.subplots(figsize=(10,5), sharex=True)

ax = plt.subplot(2,1,1)

plt.plot(pla.index * 1000, pla['PRED'], '-', label="PlanetScope SRF for Red", color='red')
plt.plot(sen.index, sen['red'], '-', label="Sentinel-2 SRF for Red", color='black')
plt.plot(mod1.index * 1000, mod1['RED'], '-', label="MODIS SRF for Red", color='gray')


ax.set_xlim([400, 1000])
ax.legend(loc = 'upper left')
plt.grid(ls = '--', alpha=.2)
labels = [item.get_text() for item in ax.get_xticklabels()]
empty_string_labels = ['']*len(labels)
ax.set_xticklabels(empty_string_labels)

plt.grid(ls = '--', alpha=.2)

ax = plt.subplot(2,1,2)

plt.plot(pla.index * 1000, pla['PNIR'], '-', label="PlanetScope SRF for NIR", color='red')
plt.plot(sen.index, sen['nir'], '-', label="Sentinel-2 SRF for NIR", color='black')
plt.plot(mod2.index * 1000, mod2['NIR'], '-', label="MODIS SRF for NIR", color='gray')

ax.set_xlim([400, 1000])
ax.legend(loc = 'upper left')
plt.grid(ls = '--', alpha=.2)
ax.set_xlabel('Wavelength (nm)')

fig.text(0.05, 0.5, 'Spectral Response Function', va='center', rotation='vertical')

plt.show()
```
